#### Changing the gender using pyttsx3

```python
import pyttsx3
from IPython.display import Audio

text = '''Im a passionate Data Scientist skilled in turning data into actionable insights. With expertise in Python, machine learning, and data visualization, I help businesses solve complex problems and make data-driven decisions. My work spans predictive modeling, NLP, and building scalable solutions to uncover hidden patterns in data.

 Lets connect to explore opportunities in data and innovation.'''

audio = pyttsx3.init()
audio.setProperty('rate', 150)
audio.setProperty('volume', 0.8)

# change the voices
voice = audio.getProperty('voices')

# 0 for male and 1 for female
audio.setProperty('voice', voice[0].id)      # for male voice   
#audio.setProperty('voice', voice[1].id)       # for female voice

# text-to speech conversion
audio.say(text)

# save the audio file
audio.save_to_file(text, 'test_male_Voice.mp3')
#audio.save_to_file(text, 'test_female_Voice.mp3')

audio.runAndWait()
```
